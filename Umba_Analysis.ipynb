```python
#Display the csv table
import pandas as pd
df = pd.read_csv('weekly_logins.csv')
df
```

```python
#Most Used Page x Weekly Logins
df = df.groupby(['Most Used Page'])
df.sum()
```

```python
#Company Name x Weekly Logins
df = df.groupby(['Company Name'])
df.sum()
```

```python
#Most Common Referrer x Weekly Logins
df = df.groupby(['Most common referrer'])
df.sum()
```

```python
#Most common IP x Weekly Logins
df = df.groupby(['Most common IP'])
df.sum()
```

```python
#Industry x Company Size x Weekly Logins
df = df.groupby(['Industry','Company Size'])
df.sum()
```

```python
#Industry x Territory x Weekly Logins
df = df.groupby(['Industry','Territory'])
df.sum()
```
